```python
frequency_table=uk_price['category'].value_counts()
frequency_table
```

```python
top_5_product=uk_price['category'].value_counts().head(5)
top_5_product
```

```python
import matplotlib.pyplot as plt  # Correct import statement
import seaborn as sns
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=top_5_product.index, y=top_5_product.values, palette="Set3")
plt.xlabel("Product Category")
plt.ylabel("Number of Listings")
plt.title("Top 5 Most Listed Product Categories on Amazon UK")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(top_5_product, labels=top_5_product.index, autopct='%1.1f%%', 
        startangle=90, colors=sns.color_palette("Set3"), wedgeprops={'edgecolor': 'black'})
plt.title("Proportion of Top 5 Product Categories on Amazon UK")
plt.show()
```

```python
#sport and outdoor product dominate the list
```

```python
mean_price=uk_price['price'].mean()
median_price=uk_price['price'].median()
mode_price=uk_price['price'].mode()
print (mean_price)
print (median_price)
print (mode_price)
```

```python
# average is 89.24 and the mode is 9.99. there is a huge difference between this two and showing average price can be higher comparing the most common one.
```

```python
variance_price=uk_price['price'].var()
std_price=uk_price['price'].std()
interquantile_price=uk_price['price'].quantile([0.25,0.50,0.75])
max_price=uk_price['price'].max()
min_price=uk_price['price'].min()
rang_price=max_price-min_price
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import numpy as np


plt.figure(figsize=(10, 6))
sns.histplot(uk_price['price'], bins=50, kde=True, color="skyblue")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Distribution of Product Prices on Amazon UK")
plt.xlim(0, uk_price['price'].quantile(0.99))  # Remove extreme outliers for better visibility
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=uk_price['price'], color="lightcoral")
plt.xlabel("Price")
plt.title("Box Plot of Product Prices on Amazon UK")
plt.xlim(0, uk_price['price'].quantile(0.99))  # Zoom into relevant price range
plt.show()
```
